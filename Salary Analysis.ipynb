```python
#function to create region name from country
def combine(countryname):
    #Asian Regions : There are six main geographical regions or subregions in Asia.Northern Asia,Western Asia,
    #Central Asia,Eastern Asia,Southern Asia,Southeast Asia.

   #Western_Asia
    if countryname in Western_Asia:
        return 'Western Asia'
    
    #Central Asia
    elif countryname in Central_Asia:
         return 'Central Asia'
        
     #East Asia  : Republic of Korea is selected by few of the respondents,so adding that too separatetly 
    elif countryname in East_Asia:
        return 'East Asia'
    
    #South Asia
    elif countryname in South_Asia:
        return 'South Asia'
    
    #Southeast Asia
    elif countryname in Southeast_Asia:
        return 'Southeast Asia'
    
    #European Regions : there are four main geographical regions or subregions in Europ.Northern Europe,Western Europe,Eastern Europe,Southern Europe
    #Northern Europe
    elif countryname in Northern_Europe:
        return 'Northern Europe'
    
    #Western Europe
    elif countryname in Western_Europe:
        return 'Western Europe'

    #Eastern Europe
    elif countryname in Eastern_Europe:
        return 'Eastern Europe'
    
    #Southern Europe
    elif countryname in Southern_Europe:
        return 'Southern Europe'
    
    #America :There are four main geographical regions or subregions in the Americas.North America,The Caribbean,Central America and South America
    #North America
    elif countryname in North_America_region:
        return 'North America'
        
    
    #Caribbean
    elif countryname in Caribbean:
        return 'Caribbean'
    
    #Central America 
    elif countryname in Central_America:
        return 'Central America'
    
    #South America
    elif countryname in South_America:
        return 'South America'
    
    
    #African Regions : There are five main geographical regions or subregions in Africa.North Africa,West Africa,Central Africa,East Africa,Southern Africa
    #North Africa
    elif countryname in North_Africa:
        
        return 'North Africa'
    
    #West Africa 
    elif countryname in West_Africa:
        return 'West Africa'
    
    #Central Africa 
    elif countryname in Central_Africa:
        return 'Central Africa'
    
    #East Africa
    elif countryname in East_Africa:
        return 'East Africa'
    
    #Southern Africa
    elif countryname in Southern_Africa:
        return 'Southern Africa'
    
   
    #Oceania:There are four main geographical regions Oceania.Australia/New Zealand,Melanesia,Micronesia,Polynesia
    #Australia/New Zealand
    elif countryname in Australia_New_Zealand:
        return 'Australia/New Zealand'
   
    #Melanesia
    elif countryname in Melanesia:
        return 'Melanesia'
    #Micronesia
    elif countryname in Micronesia:
        return 'Micronesia'
    #Polynesia
    elif countryname in Polynesia:
        return 'Polynesia'
```

```python
def continents(countryname):
    if (countryname in Northern_Europe or countryname in Western_Europe or countryname in Eastern_Europe or countryname in Southern_Europe):
        
        return 'Europe'
    
    elif(countryname in Western_Asia or countryname in Central_Asia or countryname in East_Asia or countryname in South_Asia or countryname in Southeast_Asia):
        return 'Asia'
    

    elif(countryname in North_America_region or countryname in Caribbean or countryname in Central_America):
        return 'North America'
  
    elif countryname in South_America:
        return 'South America'
    
    elif(countryname in North_Africa or countryname in West_Africa or countryname in Central_Africa or countryname in East_Africa or countryname in Southern_Africa):
        return 'Africa'
        
    elif(countryname in Australia_New_Zealand or countryname in Melanesia or countryname in Micronesia or countryname in Polynesia):
        return 'Australia'
```

# 2) Will a Professional degree or Masterâ€™s degree help me to get a good salary?

```python
Education_df=df.copy()
```

```python
#removing missing data
Education_df.dropna(subset=['EdLevel','ConvertedComp'],inplace=True)
```

```python
#what are the different education level's from the dataset?
Education_df['EdLevel'].value_counts()
```

```python
#Considering only Professional degree,Masterâ€™s degree and doctoral degree's are higher formal education
def education(degree):
    if degree in ("Professional degree (JD, MD, etc.)","Other doctoral degree (Ph.D., Ed.D., etc.)","Masterâ€™s degree (M.A., M.S., M.Eng., MBA, etc.)"):
        return 1
    else:
        return 0
```

```python
#creating region column
Education_df['Region']=Education_df['Country'].apply(combine)
```

```python
#Combining  education levels to create higher formal education column
Education_df['HigherEd']=Education_df['EdLevel'].apply(education)
Education_df['HigherEd'].mean()
```

```python
#creating two different dataframes
education_1=Education_df[Education_df['HigherEd']==1]
education_0=Education_df[Education_df['HigherEd']==0]
```

```python
education_0.head(3)
```

```python
education_1.head(3)
```

```python
#Pie Graph
labels = ["Max of Bachelors Higher Ed","Higher Formal Education"]
size = Education_df['HigherEd'].value_counts()
plt.rcParams['figure.figsize'] = (7, 5)
explode = (0, 0.1) 
g=plt.pie(size, labels=labels,explode=explode, autopct= '%1.1f%%', shadow = True, startangle = 90)
plt.title('Formal Education Distribution', fontsize = 20)
```

##### 30.1% of our respondents have higher formal education

```python
#combining higher education with regions
Region_education_0=education_0.groupby('Region')['ConvertedComp'].mean().reset_index()
Region_education_1=education_1.groupby('Region')['ConvertedComp'].mean().reset_index()
```

```python
#creating difference in earnings bar plot
from matplotlib.pyplot import *

Region_education_combined=Region_education_1.merge(Region_education_0,left_on="Region",right_on="Region",how="outer")

Region_education_combined.columns=["Region","Higher Formal Education Earnings($)","Max of Bachelors Higher Ed Earnings($)"]
Region_education_combined["Higher Formal Education Earnings($)"].fillna(0, inplace=True)


Region_education_combined['Difference in Earnings($)'] = Region_education_combined['Higher Formal Education Earnings($)'] - Region_education_combined['Max of Bachelors Higher Ed Earnings($)']

Region_education_combined.style.bar(subset=['Difference in Earnings($)'], align='mid', color=['#d65f5f', '#5fba7d'])

```

# 3) How is the average salary varied in a Small sized organizations and Established Big Companies in different regions?

```python
org_df=df.copy()
```

```python
#EDA
org_df['OrgSize'].value_counts()
```

```python
#dropping missing values
org_df.dropna(subset=['OrgSize','Employment','ConvertedComp'],inplace=True)
```

```python
def orgsize(size):
    if size in ("2 to 9 employees","10 to 19 employees","20 to 99 employees"):
        return 'Small'
    elif size in ("100 to 499 employees","500 to 999 employees"):
        return 'Medium'
    elif size in ("1,000 to 4,999 employees","5,000 to 9,999 employees","10,000 or more employee"):
        return 'Large'
   
```

```python
#considering only respondents who are employed currently
org_df=org_df[org_df['Employment']=='Employed full-time']
```

```python
org_df['Org_size']=org_df['OrgSize'].apply(orgsize)
```

```python
org_df['Org_size'].value_counts()
```

```python
#creating Region and Continent column based on country
org_df['Region']=org_df['Country'].apply(combine)
org_df['Continent']=org_df['Country'].apply(continents)
```

```python
x = org_df[org_df['Continent']=='Asia']['Region']
plt.figure(figsize = (7, 5))
g=sns.barplot(x=x,y='ConvertedComp',data=org_df,ci=None,hue='Org_size')
g.yaxis.set_ticks([25000,50000,75000,100000,125000,150000,175000,200000])

g.set(xlabel='Asia', ylabel='Salary')
g.legend(title='Organization Size',loc='upper right')
```
